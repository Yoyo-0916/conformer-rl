```python
"""
Test Dataset
"""
from pathlib import Path
from conformer_rl.datasets import QM9XYZDataset, create_qm9_dataloader

dataset_root = Path("dataset")
if not (dataset_root / "dsgdb9nsd.xyz").is_dir() and not any(dataset_root.glob("dsgdb9nsd_*.xyz")):
    dataset_root = Path("../dataset")

print("Using QM9 dataset root:", dataset_root.resolve())

dataset = QM9XYZDataset(
    root=str(dataset_root),
    target="mu",
    limit=1000,
)

data = dataset[0]

"""
Test Dataloader
"""
loader = create_qm9_dataloader(
    root=str(dataset_root),
    batch_size=32,
    shuffle=True,
    target=["homo", "lumo"],
)

batch = next(iter(loader))

"""
Peek Data
"""
print(data["x"].shape)
print(data["pos"].shape)
print(data["edge_index"].shape)
print(data["y"])
print(data["target_names"])
```
